1.Describe what an identity matrix is. What does multiplying with an identity matrix do?

An identity matrix is a square matrix, which means it has same number of columns and rows. All the elements in a diagonal of an identity matrix are always represented with "1", and all the elements apart from the diagonal is represented as "0". In simple words, we can say that whenever the column number and the row number is equal then at that poistion it is represented with "1". Whereas when the row number and column number is not equal then at those positions, it is represented with "0".

When identity matrix is multiplied by any matrix, the result is the original matrix itself. Suppose there's a matrix "A" and it is multiplied by a identity matrix "I", then the result is said to be "A" matrix itself. (A*I=A)

```python
#1.We now want to retrieve the original image. Is matrix T invertible? How would you show that it is invertible?
if T.shape[0] == T.shape[1]:
  if np.linalg.det(T)!=0:
    print("Matrix T is invertible")
  else:
    print("Matrix T is not invertible")
else:
  print("Matrix T is not square")
```

First, we need to check that the matrix T is square and also the determinant is not equal to 0. If it satisfies both the conditions, then the T is said to be invertible. Otherwise, the T matrix it is not invertible.

```python
#2.Compute the inverse and store the matrix in Tinv.
Tinv=np.linalg.inv(T)
```

```python
#3.Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.
inversed_image = np.dot(Tinv, transformed_image)
plt.imshow(inversed_image)
plt.title("Transformed Image gets back to Original Image(Row Wise)")
plt.axis('off')
plt.show()
```

* We want to perform the same operations but on the columns of the original image. How would you do the matrix algebra? Plot the scrambled image for the proof of concept.


```python
#8th Answer:

#rearange the columns and shuffle
columns=np.arange(image.shape[1])
np.random.shuffle(columns)
print(columns)

#Use your array of random column indices to create a new matrix T_col from identity_matrix
T_column = np.eye(image.shape[1])[:, columns]
print(T_column)

#Matrix multiply T_column and img_average_channel and store the result in transformed_image_columnar_wise
transformed_image_columnar_wise=np.dot(img_average_channel,T_column)
print("Shape of the Transformed Image Columnar wise->",transformed_image_columnar_wise.shape)

#Plot transformed_image_columnar_wise
plt.imshow(transformed_image_columnar_wise)
plt.title("Transformed Image Columnar Wise (Scrambled Image)")
plt.axis('off')
plt.show()

#inverse the matrix transformed_image_columnar_wise
T_column_inverse = np.linalg.inv(T_column)
transformed_inverse_image_columnar_wise = np.dot(transformed_image_columnar_wise, T_column_inverse)
plt.imshow(transformed_inverse_image_columnar_wise)
plt.title("Transformed Image gets back to Original Image (Column Wise)")
plt.axis('off')
plt.show()
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

```python
#Extra Points Answer
#Get the number of rows and columns for the original image
num_of_rows,num_of_columns=img_average_channel.shape

#rearange the rows and columns
rows = np.arange(num_of_rows)
cols = np.arange(num_of_columns)

#Shuffle the rows and columns
np.random.shuffle(rows)
np.random.shuffle(cols)

#identity matrix for both rows and columns wise
T_rows = np.eye(num_of_rows)[rows, :]
T_columns = np.eye(num_of_columns)[:, cols]

#matrix multiplication
a_matrix = np.dot(T_rows, img_average_channel)
transformed_image = np.dot(a_matrix, T_columns)
plt.imshow(transformed_image)
plt.title('Scrambled Image - Row Wise and Column Wise')
plt.axis('off')
plt.show()

#Inverse the matrix to get into its original image
T_rows_inv = np.linalg.inv(T_rows)
T_cols_inv = np.linalg.inv(T_columns)

b_matrix = np.dot(T_rows_inv, transformed_image)
transformed_inverse_image = np.dot(b_matrix, T_cols_inv)
plt.imshow(transformed_inverse_image)
plt.title('Original Image from Scrambled Image - Row Wise and Column Wise')
plt.axis('off')
plt.show()
```
